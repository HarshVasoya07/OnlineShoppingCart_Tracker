```python
# ONLINE SHOPPING CART PROJECT
class Item:
    def __init__(self, item_id, name, brand, price):
        self.item_id = item_id
        self.name = name
        self.brand = brand
        self.price = price
class Cart:
    def __init__(self):
        self.items = []
    def add_item(self, item):
        self.items.append(item)
    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
    def get_total(self):
        total = sum(item.price for item in self.items)
        return total
    def get_num_items(self):
        return len(self.items)
    
if __name__ == '__main__':
    item1 = Item(1, "iPod", "Apple", 200)
    item2 = Item(2, "Laptop", "Dell", 700)
    item3 = Item(3, "Camera", "Sony", 500)
    item4 = Item(4, "Headphones", "Bose", 250)
    item5 = Item(5, "Smartphone", "Samsung", 800)
    item6 = Item(6, "Tablet", "Microsoft", 400)
    item7 = Item(7, "Smart Watch", "Fitbit", 150)
    item8 = Item(8, "Wireless Speaker", "JBL", 300)

    myCart = Cart()

    print("Welcome to the Harsh's Tech Store!")
    print("Here are the current available items:")
    print("1. iPod - $200")
    print("2. Laptop - $700")
    print("3. Camera - $500")
    print("4. Headphones - $250")
    print("5. Smartphone - $800")
    print("6. Tablet - $400")
    print("7. Smart Watch - $150")
    print("8. Wireless Speaker - $300")

    while True:
        choice = input("Enter the item number you want to buy (or 'q' to quit): ")
        if choice == 'q':
            break
        elif choice.isdigit():
            item_number = int(choice)
            if item_number >= 1 and item_number <= 8:
                item = None
                if item_number == 1:
                    item = item1
                elif item_number == 2:
                    item = item2
                elif item_number == 3:
                    item = item3
                elif item_number == 4:
                    item = item4
                elif item_number == 5:
                    item = item5
                elif item_number == 6:
                    item = item6
                elif item_number == 7:
                    item = item7
                elif item_number == 8:
                    item = item8
                myCart.add_item(item)
                print("Item added to cart!")
            else:
                print("Invalid item number. Please try again.")
        else:
            print("Invalid input. Please try again.")

    if myCart.get_num_items() > 0:
        print("Items in your cart:")
        for item in myCart.items:
            print(f"{item.name} - ${item.price}")

        print("You have", myCart.get_num_items(), "item(s) in your cart for a total of $%.2f" % myCart.get_total())
    else:
        print("Your cart is empty.")

```

```python
#AUTO CORRECT PROJECT, ASK USER TO CONFIRM IF THE SENTENCE ENTERED IS CORRECT IF INCORRECT IT LETS THEM CORRECT IT.
class Sentence:
    def __init__(self, sentence):
        self.sentence = sentence

    def get_first_word(self):
        words = self.sentence.split()
        if words:
            return words[0]
        return None

    def get_all_words(self):
        return self.sentence.split()

    def replace(self, index, new_word):
        words = self.sentence.split()
        if 0 <= index < len(words):
            words[index] = new_word
            self.sentence = ' '.join(words)

    def __str__(self):
        return self.sentence

sentence_text = input("Enter a sentence: ")
sentence = Sentence(sentence_text)
print("First word:", sentence.get_first_word())
print("All words:", sentence.get_all_words())
index = int(input("Enter the index(number starting from 0) of the word to replace: "))
new_word = input("Enter the new word: ")
sentence.replace(index, new_word)
print("Updated sentence:", sentence)
```

```python
#ONLINE BANKING/ATM
class BankAccount:
    def __init__(self, name, password, balance=0):
        self.name = name
        self.password = password
        self.balance = balance

    def withdraw(self):
        password = input("Enter your password: ")
        if self._check_password(password):
            amount = float(input("Enter the withdrawal amount: $"))
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrawal of ${amount} successful.")
            else:
                print("Insufficient balance.")
        else:
            print("Incorrect password.")

    def deposit(self):
        password = input("Enter your password: ")
        if self._check_password(password):
            amount = float(input("Enter the deposit amount: $"))
            self.balance += amount
            print(f"Deposit of ${amount} successful.")
        else:
            print("Incorrect password.")

    def print_account_info(self):
        password = input("Enter your password: ")
        if self._check_password(password):
            print(f"Account Holder: {self.name}")
            print(f"Account Balance: ${self.balance}")
        else:
            print("Incorrect password.")

    def _check_password(self, password):
        return password == self.password
if __name__ == '__main__':
    account = BankAccount("Harsh Vasoya", "Harsh123")
    while True:
        action = input("Do you want to deposit (D) or withdraw (W)? ").upper()
        if action == "D":
            account.deposit()
            break
        elif action == "W":
            account.withdraw()
            break
        else:
            print("Invalid option. Please choose 'D' for deposit or 'W' for withdraw.")
    account.print_account_info()

```
